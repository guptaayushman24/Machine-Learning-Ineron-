```python
# Independent and Dependent Features
```

```python
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# X_train output is y_train
# X_test output is y_test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# Now we will do the feature enginnering
```

```python
# Standarize the dataset
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

Model Training

```python
# Is these a simple linear regression or multiple linear regression
# These is a multiplr linear regression problem
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression
```

```python
# fit_transform-> We are applying the formula and we are changing the data
# fit-> We are just applying the formula and we are not changing the data 
```

```python
regression.fit(X_train,y_train)
```

```python
# Print the coefficents and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# Prediction for the test data
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
# Assumptions of the linear regression
```

```python
plt.scatter(y_test,reg_pred)
plt.xlabel("Test Truth Data")
plt.ylabel("Test Predicted Data")
```

```python
# 1st Assumption:-
#  If we plot the scatter graph between the y_test and the predicted point and we get the linear data then our model is good
# 2nd Assumption:-
# error or residual = y_test-reg_pred
# The distplot of error comes near to the normal distribution curve
# 3rd Assumption:-
# Uniform distribution (Scatter Plot between reg_pred,residuals)
```

```python
residuals = y_test-reg_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind='kde')
```

```python
# Scatter plot with prediction and residual
# Uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
# Datapoints are plotted randomly they do not have any shape so it is an uniform distribution these is the one of the assumption
```

```python
# Performance Metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
# R square and adjusted R square
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
```

```python
score
```

```python
# adjusted R square
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

```python
# Adjusted R square will always be less than R square
```
